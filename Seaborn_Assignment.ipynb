Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a sample dataset
np.random.seed(42)
data = pd.DataFrame({
    'Variable1': np.random.normal(0, 1, 100),
    'Variable2': np.random.normal(5, 2, 100),
    'Category': np.random.choice(['A', 'B'], 100)
})

# Set Seaborn style
sns.set(style="whitegrid")

# 1. Histogram
plt.figure(figsize=(8, 4))
sns.histplot(data, x='Variable1', kde=True, color='skyblue')
plt.title('Histogram')
plt.show()

# 2. KDE Plot
plt.figure(figsize=(8, 4))
sns.kdeplot(data, x='Variable1', fill=True, color='coral')
plt.title('Kernel Density Estimate (KDE) Plot')
plt.show()

# 3. Scatter Plot
plt.figure(figsize=(8, 4))
sns.scatterplot(data=data, x='Variable1', y='Variable2', hue='Category', palette='Set2', s=70)
plt.title('Scatter Plot')
plt.show()

# 4. Pair Plot
plt.figure(figsize=(8, 6))
sns.pairplot(data)
plt.suptitle('Pair Plot', y=1.02)
plt.show()

# 5. Box Plot
plt.figure(figsize=(8, 4))
sns.boxplot(data=data, x='Category', y='Variable1', palette='pastel')
plt.title('Box Plot')
plt.show()
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False)

# Set plot labels and title
plt.title('Line Plot of fmri Data')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

# Show the legend
plt.legend(title='Event', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create two box plots using Seaborn
plt.figure(figsize=(12, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age', palette='pastel')
plt.title('Box Plot of Age by Pclass')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare', palette='pastel')
plt.title('Box Plot of Fare by Pclass')

# Adjust layout for better appearance
plt.tight_layout()

# Show the plots
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram using Seaborn with hue for 'cut'
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack', bins=30, palette='viridis')

# Set plot labels and title
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')

# Show the legend
plt.legend(title='Cut', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris_data = sns.load_dataset("iris")


sns.set(style="ticks")
sns.pairplot(iris_data, hue="species", palette="Set2")

plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")

flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", linewidths=.5)

plt.title('Flights Dataset Heatmap')
plt.xlabel('Year')
plt.ylabel('Month')

plt.show()
```
